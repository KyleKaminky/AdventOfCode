```python
f = open("input.txt", "r")

lines = f.readlines()

data = []

for line in lines:
    row_s = line.split()
    row = [int(x) for x in row_s[0]]
    data.append(row)
```

```python
#data = [[3,0,3,7,3],
#       [2,5,5,1,2],
#       [6,5,3,3,2],
#       [3,3,5,4,9],
#       [3,5,3,9,0]]

num_columns = len(data[0])
num_rows = len(data)

total_trees = 0
max_view = 0


for i in range(1, num_rows-1):
    for j in range(1, num_columns-1):
        height = data[i][j]
        
        seen_1 = True
        seen_2 = True
        seen_3 = True
        seen_4 = True
        
        view_1 = 0
        view_2 = 0
        view_3 = 0
        view_4 = 0
        
        kk = i-1
        while kk >=0:
            if data[kk][j] >= height:
                kk = -1
                view_1 += 1
            else:
                view_1 +=1
                kk -= 1
                
        kk = i+1
        while kk < num_rows:
            if data[kk][j] >= height:
                kk = num_rows
                view_2 += 1
            else:
                view_2 +=1
                kk += 1
                
        kk = j-1
        while kk >= 0:
            if data[i][kk] >= height:
                kk = -1
                view_3 += 1
            else:
                view_3 +=1
                kk -= 1
                
        kk = j+1
        while kk < num_columns:
            if data[i][kk] >= height:
                kk = num_columns
                view_4 += 1
            else:
                view_4 +=1
                kk += 1
        
        for kk in range(i-1, -1, -1):
            if data[kk][j] >= height:
                seen_1 = False
                
        for kk in range(i+1, num_rows):
            if data[kk][j] >= height:
                seen_2 = False
        
        for kk in range(j-1, -1, -1):
            if data[i][kk] >= height:
                seen_3 = False
              
        for kk in range(j+1, num_columns):
            if data[i][kk] >= height:
                seen_4 = False
             
        if seen_1 or seen_2 or seen_3 or seen_4:
            total_trees += 1
            
        total_view = view_1 * view_2 * view_3 * view_4
        max_view = max(max_view, total_view)
        
f.close()
print(total_trees + 2*len(data) + 2*(len(data)-2))
print(max_view)
```
